```python
import pandas as pd

# Read the dataset from a CSV file
df = pd.read_csv('actorfilms.csv')

# Display a few rows to understand the structure of the dataset
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the style for Seaborn plots (optional)
sns.set(style="whitegrid")

# Create a scatter plot for IMDb rating vs Release date
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Year', y='Rating')
plt.title('IMDb Rating vs Release Date')
plt.xticks(rotation=45)
plt.show()
```
